```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def initialize_centers(data, k):
  centers = list()
  centers = data[np.random.choice(data.shape[0], k, replace = False)]
  return centers
```

```python
def k_means(data, k):
  centers = initialize_centers(data, k)

  while True:

    distance = np.zeros((data.shape[0], k))
    for i in range(k):
      distance[: , i] = ((data - centers[i])**2).sum(axis = 1)**0.5

    closest = np.argmin(distance, axis = 1)
    old_closest = closest.copy()

    for i in range(k):
      centers[i] = data[closest == i].mean(axis = 0)


    if all(old_closest==closest):
      break

  plt.scatter(data[: , 0], data[: , 1], c = closest)
  plt.scatter(centers[: , 0], centers[: , 1], marker = '*', color = 'red')
  plt.show()
```

```python
path = '/content/data.csv'
data = np.genfromtxt(path, delimiter=',')
```

```python
k = 2

k_means(data, k)
```

```python
k = 4

k_means(data, k)
```

```python
k = 6

k_means(data, k)
```

# **Data Preparation**

```python
import random
```

```python
train = []
val = []
test = []
```

```python
def data_preparation(train_size,val_size,test_size):
  random.shuffle(data)
  for s in data:
    num = random.random()
    if num>=0 and num<=train_size:
      train.append(s)
    elif num>train_size and num<= train_size+val_size:
      val.append(s)
    else:
      test.append(s)
```

# **Classification**

```python
import math
```

```python
def euclidean_dist(v,t):
  d = 0
  for x,y in zip(v,t):
    d += (x-y)**2
  return math.sqrt(d)
```

```python
def sortfunc(d):
  return d[1]
```

```python
def kNN_classifier(k, train, val):
  distance = []
  accuracy = 0
  for v in val:
    for t in train:
      dist = euclidean_dist(v[:-1],t[:-1])
      distance.append([t,dist])
    distance.sort(key=sortfunc)
    select = distance[:k]
    label = {
        0:0,
        1:0
    }
    for s in select:
      label[s[0][-1]] += 1
    predicted_class = max(label,key=label.get)
    if v[-1]-predicted_class == 0:
      accuracy += 1
  return accuracy/len(val)
```

```python
print(kNN_classifier(10, train, val))
```

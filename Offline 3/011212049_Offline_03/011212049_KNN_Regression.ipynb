```python
train = []
val = []
test = []
```

```python
def data_preparation(train_size,val_size,test_size):
  random.shuffle(data)
  for s in data:
    num = random.random()
    if num>=0 and num<=train_size:
      train.append(s)
    elif num>train_size and num<= train_size+val_size:
      val.append(s)
    else:
      test.append(s)
```

# **Regression**

```python
import math
```

```python
def euclidean_dist(v, t):
  d = 0
  for x, y in zip(v, t):
    d += (x-y)**2
  return math.sqrt(d)
```

```python
def sortfunc(d):
  return d[1]
```

```python
def KNN_Regressor(k, train, val):
  Error = 0
  distance = []
  accuracy = 0
  for v in val:
    for t in train:
      dist = euclidean_dist(v[:-1], t[:-1])
      distance.append([t, dist])
    distance.sort(key = sortfunc)
    select = distance[:k]
    Sum = 0
    for s in select:
      Sum = Sum + s[0][1]
    Sum = Sum/k
    Error = Error + (v[-1] - Sum)**2
  return Error/len(val)
```

```python
print(KNN_Regressor(15, train, val))
```
